```python
import os
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import pandas as pd

# Define constants
IMAGE_SIZE = 150
BATCH_SIZE = 32
NUM_EPOCHS = 10
LEARNING_RATE = 0.001

# Set the paths for your training and test data
train_data_dir = '/content/train'  # Update with the correct path to your train data
test_data_dir = '/content/test'    # Update with the correct path to your test data
output_csv_file = 'predicted_labels.csv'  # Output CSV file name

# Define transformations for data augmentation and normalization
transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Custom function to filter out unwanted files
def is_valid_file(filename):
    return not any(filename.startswith(prefix) for prefix in ['.ipynb_checkpoints'])

# Load training and test datasets using ImageFolder with custom filter function
train_dataset = ImageFolder(root=train_data_dir, transform=transform, is_valid_file=is_valid_file)
test_dataset = ImageFolder(root=test_data_dir, transform=transform, is_valid_file=is_valid_file)

# Create DataLoader for training and test sets
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Define your CNN model architecture
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        # Define your layers here

    def forward(self, x):
        # Define the forward pass
        pass

# Initialize the model, loss function, and optimizer
model = CNNModel()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# Training loop
for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss = 0.0

    for i, data in enumerate(train_loader, 0):
        inputs, labels = data
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        if i % 100 == 99:
            print(f"Epoch [{epoch + 1}/{NUM_EPOCHS}], "
                  f"Batch [{i + 1}/{len(train_loader)}], "
                  f"Loss: {running_loss / 100:.4f}")
            running_loss = 0.0

# Make predictions on the test set
model.eval()
predictions = []
with torch.no_grad():
    for data in test_loader:
        inputs, _ = data
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        predictions.extend(predicted.tolist())

# Get the image names in the test set
test_image_names = [os.path.basename(f) for f, _ in test_dataset.samples]
import os
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import pandas as pd

# Define constants
IMAGE_SIZE = 150
BATCH_SIZE = 32
NUM_EPOCHS = 10
LEARNING_RATE = 0.001

# Set the paths for your training and test data
train_data_dir = '/content/train'  # Update with the correct path to your train data
test_data_dir = '/content/test'    # Update with the correct path to your test data
output_csv_file = 'predicted_labels.csv'  # Output CSV file name

# Define transformations for data augmentation and normalization
transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Load training and test datasets using ImageFolder
train_dataset = ImageFolder(root=train_data_dir, transform=transform)
test_dataset = ImageFolder(root=test_data_dir, transform=transform)

# Create DataLoader for training and test sets
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# Define the CNN model architecture
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv_layers = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.fc_layers = nn.Sequential(
            nn.Linear(128 * (IMAGE_SIZE // 8) * (IMAGE_SIZE // 8), 512),
            nn.ReLU(),
            nn.Linear(512, len(train_dataset.classes))
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x

# Initialize the model, loss function, and optimizer
model = CNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# Train the model
for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss = 0.0
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 100 == 99:  # Print every 100 mini-batches
            print(f'Epoch [{epoch + 1}/{NUM_EPOCHS}], '
                  f'Batch [{i + 1}/{len(train_loader)}], '
                  f'Loss: {running_loss / 100:.4f}')
            running_loss = 0.0

# Make predictions on the test set
model.eval()
predictions = []
true_labels = []
with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        predictions.extend(predicted.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Map predicted indices to class labels
class_labels = train_dataset.classes
predicted_labels = [class_labels[pred] for pred in predictions]

# Create a DataFrame with true labels and predicted labels
data = {'True_Label': true_labels, 'Predicted_Label': predicted_labels}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv(output_csv_file, index=False)

print(f"Predictions saved to '{output_csv_file}'")

# Create a DataFrame with image names and predicted labels
class_names = train_dataset.classes
predicted_labels = [class_names[pred] for pred in predictions]
data = {'Image': test_image_names, 'Predicted_Label': predicted_labels}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv(output_csv_file, index=False)
print(f"Predictions saved to '{output_csv_file}'")
```
