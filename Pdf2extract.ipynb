```python
import tabula

tabula.environment_info()
```

```python
name = input('Enter file name: ')
pdf_path = name+".pdf"
dfs = tabula.read_pdf(pdf_path, stream=True, pages='all')
print(dfs[0])
```

```python
import pandas as pd
import numpy as np

for i in range(len(dfs)):
  df = pd.DataFrame(dfs[i])
  cols = df.columns
  # print(df)
  # df = pd.DataFrame(dfs[])
  len_str = len(df.columns)
  df.dropna(axis=0,thresh=len_str,inplace=True)
  df.to_excel('/content/Excels/'+str(i)+'.xlsx')
```

```python
import os
!pip install xlsxwriter
def merge_excel_files(input_folder, output_file):
    all_files = os.listdir(input_folder)
    all_files_temp = []
    for i in all_files:
      x = i.strip('.xlsx')
      try:
        n = int(x)
      except:
        pass
      all_files_temp.append(n)
    all_files_temp.sort()
    all_files = []
    for i in all_files_temp:
      i = str(i)+'.xlsx'
      all_files.append(i)

    print(all_files)

    writer = pd.ExcelWriter(output_file, engine='xlsxwriter')

    for file_name in all_files:
        if file_name.endswith('.xlsx'):
            full_path = os.path.join(input_folder, file_name)
            xls = pd.ExcelFile(full_path)
            for sheet_name in xls.sheet_names:
                data = pd.read_excel(full_path, sheet_name=sheet_name)
                data.to_excel(writer, sheet_name=f"{file_name}_{sheet_name}", index=False)

    writer.save()

merge_excel_files('/content/Excels','Merged.xlsx')
```
