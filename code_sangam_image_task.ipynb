```python
import os
from PIL import Image
import tensorflow as tf
from tensorflow import keras

# List all files in the '/content/train' directory
for dirname, _, filenames in os.walk('/content/train'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# Load the dataset using image_dataset_from_directory
full_dataset = keras.utils.image_dataset_from_directory(
    directory="/content/train",
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256, 256),
    validation_split=0.2,
    subset='training',
    seed=1
)

# Validation dataset is automatically created by image_dataset_from_directory
train_ds = full_dataset

# Define the cleaning function
def cleaning(image, label):
    # Convert the image tensor to string
    image_path = tf.strings.as_string(image)

    # Check the file extension
    condition = tf.strings.regex_full_match(image_path, '.*\.(jpg|jpeg|png|bmp|gif)')

    # Apply the condition to filter the dataset
    return tf.cond(
        condition,
        lambda: (image, label),
        lambda: (tf.zeros_like(image), tf.constant(-1, dtype=tf.int32))  # Use placeholders with the correct shapes
    )

# Filter the datasets using the cleaning function
train_ds = train_ds.filter(cleaning)

# Shuffle and batch the training dataset
train_ds = train_ds.shuffle(buffer_size=1000).batch(32)

# Normalize the images
def process(image, label):
    image = tf.cast(image / 255.0, tf.float32)
    return image, label

train_ds = train_ds.map(process)
```
